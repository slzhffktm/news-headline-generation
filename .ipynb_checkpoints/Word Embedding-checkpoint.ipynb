```python
import pandas as pd
```

```python
df_X = pd.read_csv('data/X.csv')
df_y = pd.read_csv('data/y.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df_X['lemma_tokens'])
y = vectorizer.fit_transform(df_y['tokens'])
```

```python
word_list = set()
word_sets = [set(eval(l)) for l in df_X['lemma_tokens'].values]
for s in word_sets:
    word_list = word_list.union(s)
```

```python
word_dic = {i : 0 for i in word_list}
```

```python
for i in range(len(df_X['lemma_tokens'])):
    temp = eval(df_X['lemma_tokens'][i])
    for j in range (len(temp)):
        word_dic[temp[j]] += 1
```
