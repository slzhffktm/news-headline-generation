```python
import os
import requests
import json
import time

import pandas as pd
from bs4 import BeautifulSoup 
```

## Import Data

```python
def parseJson(fname):
    for line in open(fname, 'r'):
        yield eval(line)
```

```python
data = list(parseJson('./data/Sarcasm_Headlines_Dataset_v2.json'))
```

```python
theonion_data = [datum for datum in data if 'theonion.com' in datum['article_link']]
```

```python
theonion_data
```

## Scrape the Website

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
```

```python
i = 0
for news in theonion_data:
    print(i)
    i += 1
    try:
        page = requests.get(news['article_link'], headers=headers)
    except:
        theonion_data.remove(news)
        print('failed')
        continue
    soup = BeautifulSoup(page.text, 'html.parser')
    news_body = soup.find_all('p')
    if news_body == None:
        theonion_data.remove(news)
        print('failed')
        continue
    else:
        news['body'] = ' '.join([paragraph.get_text() for paragraph in news_body])
```

```python
with open('data/theonion.json', 'w') as f:
    json.dump(theonion_data, f)
```
